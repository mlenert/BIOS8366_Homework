## Question 1

The `titanic.xls` spreadsheet in the `data` directory contains data regarding the passengers on the Titanic when it sank in 1912. A recent [Kaggle competition](http://www.kaggle.com/c/titanic-gettingStarted) was based on predicting survival for passengers based on the attributes in the passenger list. 

Use scikit-learn to build both a support vector classifier and a logistic regression model to predict survival on the Titanic. Use cross-validation to assess your models, and try to tune them to improve performance.

Discuss the benefits and drawbacks of both approaches for application to such problems.

```python
# Question 1 Response
import numpy as np
import pandas as pd

from sklearn import svm
from sklearn.linear_model import LogisticRegression

data=pd.read_excel("/Users/matthewlenert/Documents/4F|Adv Stat Computing/Bios8366-master/data/titanic.xls")
# Look For Missing Data
print(data.shape)
data.isnull().sum()
```

```python
# Median imputation for variables with few missing values
median_age=data.age.median()
data.age=data.age.fillna(median_age)

median_fare=data.fare.median()
data.fare=data.fare.fillna(median_fare)

# Code dummy variable for sex
data["Is_Male"]=(data.sex=="male").astype("int")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
import random

support_vector=svm.SVC(gamma='scale')
logistic=LogisticRegression(solver="liblinear")

features=["pclass","Is_Male","age","fare","sibsp","parch"]
response=["survived"]

print("No Tunning")
svm_scores=cross_val_score(support_vector, data[features], np.ravel(data[response]), cv=10, scoring='roc_auc')
logit_scores=cross_val_score(logistic, data[features], np.ravel(data[response]), cv=10, scoring='roc_auc')
print("SVM AUC: "+str(np.mean(svm_scores))+" ± "+str(np.std(svm_scores)))
print("Logit AUC: "+str(np.mean(logit_scores))+" ± "+str(np.std(logit_scores))+"\n")


def sample_wr(population, k):
    n = len(population)
    _random, _int = random.random, int
    result = [None] * k
    for i in range(k):
        j = _int(_random() * n)
        result[i] = population[j]
    return(result)


def sampleDataFrame(x, n):
    return(x.iloc[sample_wr(x.index, n)])

print("With Tunning")
support_vector=svm.SVC(C=0.01,gamma=0.1)
bootstraps=5000
best_auc=-1
best_C=-1
best_gamma=-1
best_logit_auc=-1
best_logit_c=-1

# Implement Random Hyperparameter Search Using Out Of Bag Bootstrapping 
# Tune parameters to maximize AUROC curve on out of bag test set
for sample in range(bootstraps):
    bootstrapData=sampleDataFrame(data,data.shape[0])
    outOfBag=data[~data.index.isin(bootstrapData.index)]

    C_test=np.random.uniform(0.001,10)
    gamma_test=np.random.uniform(0.001,0.99)
    sv=svm.SVC(C=C_test,gamma=gamma_test)
    logit=LogisticRegression(C=C_test,solver="liblinear")
    logit.fit(bootstrapData[features],np.ravel(bootstrapData[response]))
    logit_test=logit.predict(outOfBag[features])
    logit_auc=metrics.roc_auc_score(np.ravel(outOfBag[response]), np.ravel(logit_test))
    
    if logit_auc>best_logit_auc:
            best_logit_auc=logit_auc
            best_logit_c=C_test
    
    try:
        sv.fit(bootstrapData[features],np.ravel(bootstrapData[response]))
        test=sv.predict(outOfBag[features])
        auc=metrics.roc_auc_score(np.ravel(outOfBag[response]), np.ravel(test))
        if auc>best_auc:
            best_auc=auc
            best_C=C_test
            best_gamma=gamma_test
    except:
        pass
    
    
support_vector=svm.SVC(C=best_C,gamma=best_gamma)  
logistic=LogisticRegression(C=best_logit_c,solver="liblinear")
svm_scores=cross_val_score(support_vector, data[features], np.ravel(data[response]), cv=10, scoring='roc_auc')
logit_scores=cross_val_score(logistic, data[features], np.ravel(data[response]), cv=10, scoring='roc_auc')
print("SVM AUC: "+str(np.mean(svm_scores))+" ± "+str(np.std(svm_scores)))
print("SVM C: "+str(best_C)+", Gamma: "+str(best_gamma)+"\n")
print("Logit AUC: "+str(np.mean(logit_scores))+" ± "+str(np.std(logit_scores)))
print("C: "+str(best_logit_c))
```

The support vector machine (SVM) appears to perform worse and have more variability in its performance than logistic regression. Tuning led to a minor improvement in the AUROC curve and a reduction in variability for the SVM, but produced neglible results for the Logistic Regression model. This parity suggests that the logistic regression model has achieved a nearly minimal amount of variance (for this particular type of model) eliminating the potential benefits of shrinkage. The use of a random search, can be more efficient at finding optimal combinations of hyper-parameters than a grid search ( http://www.jmlr.org/papers/v13/bergstra12a.html ).

The file `TNNASHVI.txt` in your data directory contains daily temperature readings for Nashville, courtesy of the [Average Daily Temperature Archive](http://academic.udayton.edu/kissock/http/Weather/). This data, as one would expect, oscillates annually. Using PyMC3, use a Gaussian process to fit a non-parametric regression model to this data, choosing an appropriate covariance function. Plot 10 regression lines drawn from your process.

```python
%matplotlib inline
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import matplotlib.cm as cmap

daily_temps = pd.read_table("/Users/matthewlenert/Documents/4F|Adv Stat Computing/Bios8366-master/data/TNNASHVI.txt", sep='\s+', 
                            names=['month','day','year','temp'], na_values=-99)

```

```python
daily_temps.temp.plot(style='b.', figsize=(10,6), grid=False)
```

```python
# Question 2
import numpy as np
import scipy as sp
import pandas as pd 
import pymc3 as pm
import warnings
warnings.filterwarnings("ignore", module="theano")
warnings.filterwarnings("ignore", module="mkl_fft")
warnings.filterwarnings("ignore", module="matplotlib")

np.random.seed(1776)
X=daily_temps.dropna().index.values[:, None]
Y=daily_temps.dropna().temp.values

with pm.Model() as bna_temperature_model:
    l = pm.Gamma("l", alpha=2, beta=1)
    tau = pm.HalfCauchy("tau", beta=5)
    # Using Matern for flexability
    # Attempted using cosine, but sampling was significantly was slower
    cov = tau**2 * pm.gp.cov.Matern52(1, l)
    m = pm.gp.mean.Constant(daily_temps.mean())
    gp = pm.gp.MarginalSparse(cov_func=cov, approx="FITC")
    Xu = pm.gp.util.kmeans_inducing_points(45, X)
    
```

```python
with bna_temperature_model:
    
    σ = pm.HalfCauchy("σ", beta=5)
    obs = gp.marginal_likelihood("obs", X=X, Xu=Xu, y=Y, noise=σ)
    
    trace = pm.sample(1000, chains=2)
    pm.traceplot(trace)
    pm.energyplot(trace)
    
print("Gelman-Rubin: "+str(pm.gelman_rubin(trace)))
    
```

The Gelman-Rubin statistics do not show evidence of divergence in the chains. The estimated parameters appear stable over multiple chains in the trace plots and do not exhibit signs of biased motion or drifting toward an asymptote. The energy plot does not show evidence of poor exploration of the distribution by the NUTS algorithm. Failing to see evidence of bias in estimating the parameters, I proceed to sample from the model. 

```python
with bna_temperature_model:
    predicted_t = gp.conditional("predicted_t", X, pred_noise=True)
    
with bna_temperature_model:
    pred_samples = pm.sample_ppc(trace, vars=[predicted_t], samples=10)
```

```python
fig = plt.figure(figsize=(12,5)); ax = fig.gca()
  
x = X.flatten()
# plot the samples from the gp posterior with samples and shading
from pymc3.gp.util import plot_gp_dist
plot_gp_dist(ax, pred_samples["predicted_t"], x)

# plot the data and the true latent function
plt.plot(x, Y, 'ok', ms=3, alpha=0.5, label="Observed data")
plt.plot(Xu, 10*np.ones(Xu.shape[0]), "co", ms=10, label="Inducing point locations")

# axis labels and title
plt.xlabel("Days From First Measurement")
plt.ylabel("Temperature (F)")
plt.title("Posterior distribution over a function of the day at the observed values"); plt.legend();\
```

To produce a resonable estimate of the tight (frequent) periodicity of the data, many inducing locations were required. I attempted fitting the model with 15, 20, 40, and 45 inducing locations. Uing 45 produced the best results, pictured above. 

## Question 3

Fit a series of random-forest classifiers to the Wisconsin breast cancer dataset (`wisconsin_breast_cancer.csv`), to explore the sensitivity to the parameters `max_features`, the number of variables considered for splitting at each step, `max_depth`, the maximum depth of the tree, and `n_estimators`, the number of trees in the forest. Use apprpriate metrics of performance, and include plots against a suitably-chosen range of values for these parameters.

Dataset description: Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image. Ten real-valued features are computed for each cell nucleus:

- `radius` (mean of distances from center to points on the perimeter) 
- `texture` (standard deviation of gray-scale values)

- `symmetry`

The outcome to be predicted is tumor type (M = malignant, B = benign).

```python
# Question 3
import numpy as np
import pandas as pd

data=pd.read_csv("/Users/matthewlenert/Documents/4F|Adv Stat Computing/Bios8366-master/data/wisconsin_breast_cancer.csv")

# Explore Missing Data
count=0
print_count=0
for missing_count in data.isnull().sum():
    if missing_count > 0:
        print(data.columns.values[count]+": "+str(missing_count))
        print_count+=1
    count+=1
if print_count==0:
    print("No missing values\n")

# Get data types
data.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import random

# Gather features by datatype
g = data.columns.to_series().groupby(data.dtypes).groups
g = {k.name: v for k, v in g.items()}
features=g['float64']

# Code outcome as numeric
data["label"]=data.diagnosis=="M"
response=["label"]

rf=RandomForestClassifier()

bootstraps=2000
best_depth=-1
best_forest_size=-1
best_num_features=-1
best_f1=-1

log={"depth":[],"forest_size":[],"num_features":[],"precision":[],"recall":[],"f1":[]}


def sample_wr(population, k):
    n = len(population)
    _random, _int = random.random, int
    result = [None] * k
    for i in range(k):
        j = _int(_random() * n)
        result[i] = population[j]
    return(result)


def sampleDataFrame(x, n):
    return(x.iloc[sample_wr(x.index, n)])


# Implement Random Hyperparameter Search Using Out Of Bag Bootstrapping 
for sample in range(bootstraps):
    bootstrapData=sampleDataFrame(data,data.shape[0])
    outOfBag=data[~data.index.isin(bootstrapData.index)]
    
    # Randomly Select Parameters 
    depth=np.random.randint(1,len(features))
    size=np.random.randint(10,250)
    feature_count=np.random.randint(1,len(features))
    
    # Log Values
    log["depth"].append(depth)
    log["forest_size"].append(size)
    log["num_features"].append(feature_count)
    
    # Fit Model on Bootstrap
    rf=RandomForestClassifier(n_estimators=size,max_depth=depth,max_features=feature_count)
    rf.fit(bootstrapData[features],np.ravel(bootstrapData[response]))
    
    # Test on Out of Bag
    rf_test=rf.predict(outOfBag[features])
    
    rf_precision=metrics.precision_score(np.ravel(outOfBag[response]), np.ravel(rf_test))
    rf_recall=metrics.recall_score(np.ravel(outOfBag[response]), np.ravel(rf_test))
    rf_f1=metrics.f1_score(np.ravel(outOfBag[response]), np.ravel(rf_test))
    
    # Log Results
    log["precision"].append(rf_precision)
    log["recall"].append(rf_recall)
    log["f1"].append(rf_f1)
    
    
    if rf_f1>best_f1:
        best_f1=rf_f1
        best_depth=depth
        best_forest_size=size
        best_num_features=feature_count
```
