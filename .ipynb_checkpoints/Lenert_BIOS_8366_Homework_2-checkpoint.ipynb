# Question 1

Suppose $y$ has a binomial distribution with parameters $n$ and $p$, and we are interested in the log-odds value $\theta = \log(p/(1-p))$. Our prior for $\theta$ is that $\theta \sim N(\mu, \sigma^2)$. It follows that the posterior density of $\theta$ is given, up to a proportionality constant, by:

$$\pi(\theta | y) \propto \frac{\exp(y\theta)}{(1 + exp(\theta))^n} \exp\left[\frac{-(\theta - \mu)^2}{2\sigma^2}\right]$$

For example, suppose we are interested in learning about the probability that a possibly-biased coin lands heads when tossed. *A priori* we believe that the coin is fair, so we assign $\theta$ a $N(0,.25)$ prior. We toss the coin $n = 5$ times and obtain $y = 5$ heads.

1. Using a normal approximation to the posterior density, compute the probability that the coin is biased toward heads (i.e., that θ is positive).
2. Using the prior density as a proposal density, design a rejection algorithm for sampling from the posterior distribution. Using simulated draws from your algorithm, approximate the probability that the coin is biased toward heads.
3. Using the prior density as a proposal density, simulate values from the posterior distribution using the SIR algorithm. Approximate the probability that the coin is biased toward heads.

```python
# Question 1.1

from scipy.optimize import minimize
from scipy.stats import norm

def log_posterior(theta,k,n,mu,sigma):
    post=(k*theta)
    post-=np.log((1+np.exp(theta)))*n
    post-=((theta-mu)**2)/(2*sigma**2)
    return(-1*post)

def normal_approx_for_theta(k,n,mu_prior,sigma_prior):
    # bfgs on the log posterior to optimize theta
    opt = minimize(log_posterior, np.random.normal(mu_prior,sigma_prior), method='L-BFGS-B',args=(k,n,mu_prior,sigma_prior))
    mode = opt.x
    var = opt.hess_inv.todense()
    # Use cdf to calulate P(theta > 0)
    print("The normal approximation estimates a "+str(round(1-norm.cdf(0,*mode,*var)[0],2)*100)+ "% probability that the coin is biased toward heads")
    
normal_approx_for_theta(5,5,0,0.5)
```

```python
# Question 1.2

# Posterior PDF 
def posterior(theta,k,n,mu,sigma):
    post=np.exp(k*theta)/(1+np.exp(theta))**n
    post=post*np.exp(-1*(theta-mu)**2/(2*sigma**2))
    return(post)

def rejection_sample(desired_pdf,desired_samples,k,n,mu,sigma):
    # Initialize variables
    samples=np.zeros(desired_samples)
    sample_index=0
    c=1
    biggestDiff=0
    
    # Continue until desired # of samples acquired
    while sample_index<desired_samples:
        trials=min(desired_samples,2*desired_samples-sample_index)
        
        # Prior is proposed distriubtion
        g=np.random.normal(mu,sigma,trials)
        u=np.random.uniform(size=trials)
        for i in range(trials):
            # Accept/Reject algorithm
            ratio=desired_pdf(g[i],k,n,mu,sigma)/(c*norm.pdf(g[i],mu,sigma))

            # Wait until we find a good candidate supremum of the ratio
            # of the desired pdf to the known pdf
            if c==1:
                
                #Acceptance criteria
                if u[i] <= ratio:
                    samples[sample_index]=g[i]
                    sample_index+=1
                    
                    if sample_index>=desired_samples:
                        break
                        
            # Check if supremum needs adjustment            
            if ratio > 1:
                if ratio > biggestDiff:
                    biggestDiff=ratio
            if ratio < 0.01:
                biggestDiff=biggestDiff/10
            
                
        # Adjust supremum if a better value is found
        c=max(biggestDiff,c)
    
    return(samples)


def getEmpiricalCumulativProbability(sample,x):
    return(np.sum(sample<=x)/len(sample))

prob=1-getEmpiricalCumulativProbability(rejection_sample(posterior,1000,5,5,0,0.5),0)
print("Rejection sampling from the prior estimates an "+str(round(prob,2)*100)+ "% probability that the coin is biased toward heads")
    
```

```python
# Question 1.3

def SIR_sample(desired_pdf,desired_samples,k,n,mu,sigma):
    # Generate candidates
    theta=np.random.normal(mu,sigma,desired_samples)
    
    # Find probabilities under posterior
    f_theta=np.array([(posterior(t,k,n,mu,sigma)) for t in theta])
    
    # Find probabilities under known distrubtion
    q_theta=np.array([norm.pdf(t,mu,sigma) for t in theta])
    
    # Calculate relative weights
    w = f_theta / q_theta / max(f_theta - q_theta)
    
    # Normalize weights
    p_sir = w/w.sum()
    
    # Sample from candidates based on weights
    theta_sir = theta[np.random.choice(range(len(theta)), size=desired_samples, p=p_sir)]
    
    return(theta_sir)


prob=1-getEmpiricalCumulativProbability(SIR_sample(posterior,10000,5,5,0,0.5),0)
print("SIR from the prior estimates a "+str(round(prob,1)*100)+ "% probability that the coin is biased toward heads")
   
```

# Question 2

The goal of this problem is to investigate the role of the proposal distribution in a Metropolis-Hastings algorithm designed to simulate from the posterior distribution of the mixture parameter $\delta$. 

1. Simulate 200 realizations from the mixture distribution:
    $$y_i \sim \delta N(7, 0.5^2) + (1-\delta) N(10, 0.5^2)$$
    with $\delta = 0.7$. Plot a histogram of these data. 
2. Implement a random walk M-H algorithm with proposal $\delta^{\prime} = \delta^{(i)} + \epsilon$ with $\epsilon \sim Unif(-1,1)$. 
3. Reparameterize the problem letting $U = \log\left[\frac{\delta}{1 - \delta}\right]$ and $u^{\prime} = u^{(i)} + \epsilon$. Implement a random walk chain in U-space. 
4. Compare the estimates and convergence behavior of the three algorithms.

In part (1), you are asked to simulate data from a distribution with $\delta$ known. For parts (2)–(3), assume $\delta$ is unknown with prior $\delta \sim Unif( 0,1)$. For parts (2)–(3), provide an appropriate plot and a table summarizing the output of the algorithm. 

To facilitate comparisons, use the same number of iterations, random seed, starting values, and burn-in period for all implementations of the algorithm. 

```python
# Question 2.1
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt


def genMixedSamples(samples,delta,m1,s1,m2,s2):
    indicator=np.random.uniform(size=samples)
    mix1=np.random.normal(m1,s1,samples)
    mix2=np.random.normal(m2,s2,samples)
    
    for i in range(samples):
        if indicator[i] > delta:
            mix1[i]=mix2[i]
    return(mix1)

    
mixtureData=genMixedSamples(200,0.7,7,0.5,10,0.5)

plt.hist(mixtureData,bins=20)
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.title('Histogram of Simulated Mixture Data')
plt.show()
```

```python
# Question 2.2

init_val1=0.95
init_val2=0.05

def calc_mixture_posterior(delta,m1,s1,m2,s2,y):
    log_prob=sum(np.log((delta*norm.pdf(y,m1,s1)+(1-delta)*norm.pdf(y,m2,s2))))
    return(log_prob)

def mh_walk(p):
    return(np.random.uniform())

def metropolis(desired_iterations,burn_in,posterior,walk_function,initial_value,m1,s1,m2,s2,y):
    n_iterations=desired_iterations+burn_in
    
    # Initialize trace for parameters
    trace = np.empty((n_iterations+1, 1))
    
    # Set initial values
    trace[0] = initial_value
        
    # Calculate joint posterior for initial values
    current_log_prob = posterior(trace[0],m1,s1,m2,s2,y)
    
    # Initialize acceptance counts
    accepted = np.zeros(n_iterations+1)
    
    for i in range(n_iterations):
    
        if not i%1000: print('Iteration %i' % i)
    
        # Grab current parameter values
        current_params = trace[i]

    
        # Get current value for parameter
        p = trace[i]

        
        
        p = walk_function(p)
        

        # Calculate log posterior with proposed value
        proposed_log_prob = posterior(p,m1,s1,m2,s2,y)
    
        
        # Log-acceptance rate
        alpha = proposed_log_prob - current_log_prob
        
        # Sample a uniform random variate
        u = np.log(np.random.uniform())

        # Test proposed value
        if u < alpha:
            # Accept
            trace[i+1] = p
            current_log_prob = proposed_log_prob
            accepted[i+1] = 1
        else:
            # Reject
            trace[i+1] = trace[i]
            accepted[i+1] = 0
                
    return (trace[burn_in:], accepted[burn_in:])


trace1,accept1 = metropolis(10000,250,calc_mixture_posterior,mh_walk,init_val1,7,0.5,10,0.5,mixtureData)
trace2, accept2 = metropolis(10000,250,calc_mixture_posterior,mh_walk,init_val2,7,0.5,10,0.5,mixtureData)

# 2.1 5% (mixture from 2 distributions)
# 2.3 30% 
# to compare acceptance rate, trace (stationary distribution of the trace after burn in)
# 
```

```python
print("M-H acceptance rate was "+str(round(np.mean(accept1),2)*100)+"% for chain 1 and "+str(round(np.mean(accept2),2)*100)+"% for chain 2")
fig, axes = plt.subplots(1, 2,figsize=(10, 3))
axes[0].set_title("Chain 1")
axes[0].plot(trace1)
axes[0].set_ylim(0,1)
axes[0].set_ylabel(r"$\delta$")
axes[0].set_xlabel("Iteration")
axes[1].hist(trace1)
axes[1].set_ylabel("Frequency")
axes[1].set_xlabel(r"$\delta$")

fig, axes = plt.subplots(1, 2,figsize=(10, 3))
axes[0].set_title("Chain 2")
axes[0].plot(trace2)
axes[0].set_ylim(0,1)
axes[0].set_ylabel(r"$\delta$")
axes[0].set_xlabel("Iteration")
axes[1].hist(trace2)
axes[1].set_ylabel("Frequency")
axes[1].set_xlabel(r"$\delta$")
```

```python
# Question 2.3

# Re-parameterize posterior
def calc_mixture_u_posterior(u,m1,s1,m2,s2,y):
    d=(np.exp(u)/(1+np.exp(u)))
    log_prob=sum(np.log((d*norm.pdf(y,m1,s1)+(1-d)*norm.pdf(y,m2,s2))))
    return(log_prob)

# Change update function to traverse the real number line
def u_walk(p):
    return(p+np.random.uniform(-1,1))

trace1, accept1 = metropolis(10000,250,calc_mixture_u_posterior,u_walk,init_val1,7,0.5,10,0.5,mixtureData)
trace2, accept2 = metropolis(10000,250,calc_mixture_u_posterior,u_walk,init_val2,7,0.5,10,0.5,mixtureData)
```

```python
print("U transformed M-H acceptance rate was "+str(round(np.mean(accept1),2)*100)+"% for chain 1 and "+str(round(np.mean(accept2),2)*100)+"% for chain 2")
fig, axes = plt.subplots(1, 2,figsize=(10, 3))
axes[0].set_title("Chain 1")
axes[0].plot(trace1)
axes[0].set_ylim(0,1)
axes[0].set_ylabel(r"u")
axes[0].set_xlabel("Iteration")
axes[1].hist(trace1)
axes[1].set_ylabel("Frequency")
axes[1].set_xlabel(r"u")

fig, axes = plt.subplots(1, 2,figsize=(10, 3))
axes[0].set_title("Chain 2")
axes[0].plot(trace2)
axes[0].set_ylim(0,1)
axes[0].set_ylabel(r"u")
axes[0].set_xlabel("Iteration")
axes[1].hist(trace2)
axes[1].set_ylabel("Frequency")
axes[1].set_xlabel(r"u")
```

Both versions of the MH algorithm appear to have converged in chains that began on opposite sides of the parameter space. Niether versions of the MH algorithm demonstrates strong signs of autocorrelation. They both appearred to produce stationary distributions centered on the estimate for delta. The plain MH algorithm experienced much tighter convergence than the u-space tranformation. The u-space version had double the acceptance rate and was generally more variable than the plain MH algorithm. The increased variability in u-space enabled the algorithm to produce superior  acceptance values. 

# Question 3

Carlin (1992) considers a Bayesian approach to meta-analysis, and includes the following examples of 22 trials of beta-blockers to prevent mortality after myocardial infarction. These data are given below.

In one possible random effects model we assume the true baseline mean (on a log-odds scale) $m_i$ in a trial $i$
is drawn from some population distribution. Let $r^C_i$ denote number of events in the control group in trial $i$, and $r^T_i$ denote events under active treatment in trial $i$. Our model is:

$$\begin{aligned}
r^C_i &\sim \text{Binomial}\left(p^C_i, n^C_i\right) \\
r^T_i &\sim \text{Binomial}\left(p^T_i, n^T_i\right) \\
\text{logit}\left(p^C_i\right) &= \mu_i \\
\text{logit}\left(p^T_i\right) &= \mu_i + \delta \\
\mu_i &\sim \text{Normal}(m, s).
\end{aligned}$$

In this case, we want to make inferences about the population effect $m$, and the predictive distribution for the effect $\delta_{\text{new}}$ in a new trial. 

This particular model uses a random effect for the population mean, and a fixed effect for the treatment effect. There are 3 other models you could fit to represent all possible combinations of fixed or random effects for these two parameters.

Build all 4 models to estimate the treatment effect in PyMC3 and 

1. use convergence diagnostics to check for convergence in each model 
2. use posterior predictive checks to compare the fit of the models
3. use an appropriate metric to compare the models as approximations of the true generating model

Which model would you select and why?

```python
r_t_obs = [3, 7, 5, 102, 28, 4, 98, 60, 25, 138, 64, 45, 9, 57, 25, 33, 28, 8, 6, 32, 27, 22]
n_t_obs = [38, 114, 69, 1533, 355, 59, 945, 632, 278,1916, 873, 263, 291, 858, 154, 207, 251, 151, 174, 209, 391, 680]
r_c_obs = [3, 14, 11, 127, 27, 6, 152, 48, 37, 188, 52, 47, 16, 45, 31, 38, 12, 6, 3, 40, 43, 39]
n_c_obs = [39, 116, 93, 1520, 365, 52, 939, 471, 282, 1921, 583, 266, 293, 883, 147, 213, 122, 154, 134, 218, 364, 674]
N = len(n_c_obs)
```

```python

with pm.Model() as all_fixed:
    
    m = pm.Normal('m', mu=0.,tau=0.0001)
    d = pm.Beta('d', 1, 1)
    p_c = pm.invlogit(m)
    p_t = pm.invlogit(m+d)
    r_c = pm.Binomial('r_c', p=p_c, observed=r_c_obs, n=n_c_obs)
    r_t = pm.Binomial('r_t', p=p_t, observed=r_t_obs, n=n_t_obs)
    all_fixed_trace = pm.sample(1000, tune=2000, nuts_kwargs={'target_accept': .95})

```

```python
with pm.Model() as random_mu:
    
    m_i = pm.Normal('m_i', mu=0.,tau=0.0001)
    s = pm.HalfCauchy('s', 5)
    m = pm.Normal('m',mu=m_i,sd=s, shape=N)
    d = pm.Beta('d', 1, 1)
    p_c = pm.invlogit(m[range(N)])
    p_t = pm.invlogit(m[range(N)]+d)
    r_c = pm.Binomial('r_c', p=p_c, observed=r_c_obs, n=n_c_obs)
    r_t = pm.Binomial('r_t', p=p_t, observed=r_t_obs, n=n_t_obs)
    random_mu_trace = pm.sample(1000, tune=2000, nuts_kwargs={'target_accept': .95})
```

```python
with pm.Model() as random_d:
    
    m = pm.Normal('m', mu=0.,tau=0.0001)
    a = pm.Gamma('a',alpha=0.001,beta=0.001)
    b = pm.Gamma('b',alpha=0.001,beta=0.001)
    d = pm.Beta('d', alpha=a, beta=b, shape=N)
    p_c = pm.invlogit(m)
    p_t = pm.invlogit(m+d[range(N)])
    r_c = pm.Binomial('r_c', p=p_c, observed=r_c_obs, n=n_c_obs)
    r_t = pm.Binomial('r_t', p=p_t, observed=r_t_obs, n=n_t_obs)
    random_d_trace = pm.sample(1000, tune=2500, nuts_kwargs={'target_accept': .90})
```

```python
with pm.Model() as all_random:
    
    m_i = pm.Normal('m_i', mu=0.,tau=0.0001)
    s = pm.HalfCauchy('s', 5)
    m = pm.Normal('m',mu=m_i,sd=s, shape=N)
    a = pm.Gamma('a',alpha=0.001,beta=0.001)
    b = pm.Gamma('b',alpha=0.001,beta=0.001)
    d = pm.Beta('d', alpha=a, beta=b, shape=N)
    p_c = pm.invlogit(m)
    p_t = pm.invlogit(m+d[range(N)])
    r_c = pm.Binomial('r_c', p=p_c, observed=r_c_obs, n=n_c_obs)
    r_t = pm.Binomial('r_t', p=p_t, observed=r_t_obs, n=n_t_obs)
    all_random_trace = pm.sample(1000, tune=2500, nuts_kwargs={'target_accept': .95})
```
